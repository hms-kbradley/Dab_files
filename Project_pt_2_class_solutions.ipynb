```python
#!pip install requests 
#!pip install pandas 
#!pip install io 
import requests 
import pandas as pd 
from io import StringIO 
import matplotlib.pyplot as plt 
```

```python
#make a function
def get_dataframe_from_google_sheet(workbook_id, worksheet_id):

    url = f'https://docs.google.com/spreadsheets/d/{workbook_id}/export?format=csv&gid={worksheet_id}'  
    response = requests.get(url)
    data = response.content
    df_data = pd.read_csv(StringIO(data.decode('utf-8')))
    df = pd.DataFrame(df_data)

    return df
```

```python
product = get_dataframe_from_google_sheet('1m5QiwYV1Jxen43tJR_moqdWm1POYwIQBpo2_m6Io6sg', '268662506') 
customer = get_dataframe_from_google_sheet('1m5QiwYV1Jxen43tJR_moqdWm1POYwIQBpo2_m6Io6sg', '1184288165')
segment =get_dataframe_from_google_sheet('1m5QiwYV1Jxen43tJR_moqdWm1POYwIQBpo2_m6Io6sg', '568436272')
sales = get_dataframe_from_google_sheet('1m5QiwYV1Jxen43tJR_moqdWm1POYwIQBpo2_m6Io6sg', '2071518297')  
```

```python
#pandas.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, 
#sort=False, suffixes=('_x', '_y'), copy=None, indicator=False, validate=None)[source]
sales_customer_joined = pd.merge(sales, customer, how = 'inner' , left_on = 'CustomerID', right_on = 'CustomerID')
add_product = pd.merge(sales_customer_joined, product, how = 'left', on='ProductID')
final_df = pd.merge(add_product, segment, how = 'left', on='CustomerID')
```

```python
def summary_crosstab_SaleAmount(ColumnName, Agg_Name):
    df = pd.crosstab(
    final_df[ColumnName],
    columns = Agg_Name,
    values = final_df['SaleAmount'],
    aggfunc = Agg_Name).reset_index()

    return df

summary_crosstab_SaleAmount('FrequentBuyers', 'sum')
```
